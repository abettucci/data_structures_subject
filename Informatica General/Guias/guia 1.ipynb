```python
def leerNumMayor1():
  n = int (input("Ingrese un numero:"))
  while n <= 1:
    n = int (input("Ingrese un numero mayor que 1:"))
  return n
def mcd(a,b):
  maximo = 1
  div = 1
  while div <= a and div <= b:
    if a % div == 0 and b % div == 0:
      maximo = div
    div +=1
 return maximo

def sonCoprimos(a,b):
 return mcd(a,b)==1

#Programa principal
a = leerNumMayor1()
b = leerNumMayor1()
while not sonCoprimos(a,b):
 a = leerNumMayor1()
 b = leerNumMayor1()
print (a, " y ", b, " son coprimos!")
```
