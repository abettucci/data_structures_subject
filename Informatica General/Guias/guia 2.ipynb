```python
#primero se lee la cadena
cadena = input('Ingrese un nmero en hexadecimal:')
#luego se valida que tenga unicamente los simbolos indicados
esHexa = True
for simbolo in cadena:
  if not((simbolo >='0' and simbolo <='9') or (simbolo >='a' and simbolo <='f')):
    esHexa = False

#luego se transforma a base 10
if esHexa:
  numero=0
  for simbolo in cadena:
    numero*=16
    numero+=int(simbolo,base=16)
print("El numero es:", numero)
```

```python
cadena = input('Ingrese un texto: ')
k = int(input('Ingrese un valor para la rotacin:'))
cifrada = ''
for simbolo in cadena:
  diferencia = ord(simbolo)-ord('a')
  ubicacion = (diferencia + k)%26 + ord('a')
  cifrada = cifrada + chr(ubicacion)
print('La cadena cifrada es: ', cifrada)
```
