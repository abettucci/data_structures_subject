```python
def ingresarLista (lista, tam):
  for i in range(tam):
    print(i,end='. ')
    elemento = int(input("Ingrese un numero para la lista:"))
    lista = lista + [elemento]
  return lista

def comparaListas(lista1,lista2):
  i = 0
  tamA = len(lista1)
  tamB = len(lista2)
  dif = 0
  while (i < tamA and i < tamB) and dif == 0:
    dif = lista1[i]-lista2[i]
    #si la diferencia es cero son iguales
    #si la diferencia es negativa la 1 es menor
    #si la diferencia es positiva la 1 es mayor
    i+=1
  if dif == 0: #entonces salio porque alguna lista se termino.
    if i < tamA and i == tamB:#se termino la lista 2 y la 1 no.
      dif = 1
    if i == tamA and i < tamB:#se termino la lista 1 y la 2 no.
      dif = -1
  return dif

#Programa principal
print("Lista 1:")
a = []
tamA = int(input("Ingrese tam:"))
a = ingresarLista(a, tamA)
print("Lista 2:")
b = []
tamB = int(input("Ingrese tam:"))
b = ingresarLista(b, tamB)
print(a)
print(b)
if comparaListas(a, b)<0:
  print("La primera es menor que la segunda")
```

Escribir un programa que, dada una lista de 20 nmeros enteros cargados aleatoriamente, los ordene de mayor a menor.

```python
def listacargar(tam):
  lista=[]
  for i in range(tam):
    numero = randint(-40,40)
    lista=lista+[numero]
  return lista

def ordeno(lista):
  dim = len(lista)
  for i in range(dim-1):
    for j in range(i+1,dim):
      if lista[i] < lista[j]:
        aux = lista[j]
        lista[j] = lista[i]
        lista[i] = aux
  return lista

tam=int(input("ingrese tamao de la lista"))
print("la lista es:")
lista=listacargar(tam)
print(lista)
rta=ordeno(lista)
print(rta)
```
